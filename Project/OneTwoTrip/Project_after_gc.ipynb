###  Lets' look at Field9, Field24, Field28 and Field15

```python
def get_hist(column):
    labels = list(range(df_train[column].min(), df_train[column].max()+1))
    returned = (df_train[df_train.goal1 == 1][column].value_counts()/df_train[df_train.goal1 == 1][column].value_counts().sum()).values.tolist()
    not_returned = (df_train[df_train.goal1 == 0][column].value_counts()/df_train[df_train.goal1 == 0][column].value_counts().sum()).values.tolist()
    returned += [0]*(len(labels)-len(returned))
    not_returned += [0]*(len(labels)-len(not_returned))
    x = np.arange(len(labels))
    return returned, not_returned, x, labels
```

```python
def plot(columns, n = 2):
    fig, axes = plt.subplots(nrows=2, ncols=n, figsize=(15, 9))
    for i, column in enumerate(columns):
        y_ = i % n
        x_ = i // n
        returned, not_returned, x, labels = get_hist(column)
        width = 0.3

        axes[x_, y_].bar(x - width/2, returned, width, label='returned')
        axes[x_, y_].bar(x + width/2, not_returned, width, label='not returned')
        axes[x_, y_].set_ylabel('')
        axes[x_, y_].set_xlabel(column)
        if labels[0] == 1:
            axes[x_, y_].set_xticks(x-1)
        else:
            axes[x_, y_].set_xticks(x)
        axes[x_, y_].set_xticklabels(x)
        axes[x_, y_].legend()

    plt.show()
```

```python
def for_pretty_presentation():
    fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 12))

    sns.scatterplot(df_train.field24, df_train.field15, ax= axes[0, 0])
    axes[0, 0].set_xticks(range(1,19))
    axes[0, 0].set_yticks(range(1,19))

    sns.scatterplot(df_train.field9, df_train.field15, ax= axes[0, 1])
    axes[0, 1].set_xticks(range(-1,19))
    axes[0, 1].set_yticks(range(-1,19))

    sns.scatterplot(df_train.field28, df_train.field15, ax= axes[1, 0])
    axes[1, 0].set_xticks(range(-1,19))
    axes[1, 0].set_yticks(range(-1,19))

    sns.scatterplot(df_train.field28+df_train.field24+df_train.field9, df_train.field15, ax= axes[1, 1])
    axes[1, 1].set_xticks(range(1,19))
    axes[1, 1].set_yticks(range(1,19))
    axes[1, 1].set_xlabel('field28+field24+field9')

    plt.show()
```

### Let's preprocess our data 

```python
from sklearn.preprocessing import MinMaxScaler
import warnings
warnings.filterwarnings('ignore')
from sklearn.preprocessing import OneHotEncoder
```

```python
class Preprocesser:
    X = None
    y = None
    def fit(self,X_):
        self.X = X_
        self.X['field24'] = self.X['field24'].apply(lambda x: 6 if x > 5 else x)
        self.X['number_of_tickets_to_date'] = self.X.sort_values(['field21','field2']).groupby('userid')['field15'].cumsum()

    def transform(self,X_):
        tmp = X_.copy()
        #transform columns with 1,2 values to 0,1
        for column in ['field21', 'field7']: 
            tmp[column] = tmp[column].apply(lambda x: 0 if x == 2 else x)
        #create new numerical feature: number of tickets bought by the user at the current moment of time
        tmp['number_of_tickets_to_date'] = tmp.sort_values(['field21','field2']).groupby('userid')['field15'].cumsum()
        #scale float columns
        float_features = ['field1', 'field14', 'number_of_tickets_to_date','field16']
        scaler = MinMaxScaler()
        scaler.fit(self.X[float_features])
        tmp[float_features] = scaler.transform(tmp[float_features])
        # drop irrelevant features
        features_to_drop = ["field0", "field4", "field6", "field12",
                            "field13","field17","field19","field22","field25",
                            "field26","field27","field28","indicator_goal21",
                            "indicator_goal22","indicator_goal23",
                            "goal23","goal25"]
        tmp.drop(features_to_drop, inplace=True, axis=1)
        tmp.drop(columns = 'userid', inplace = True)
        #some binary prepro, according to the feature meaning
        tmp['field9'] = tmp['field9'].apply(lambda x: not x)
        for column in ["field15","field24"]:
            tmp[column] = tmp[column].apply(lambda x: 6 if x > 5 else x)
        tmp['field18'] = tmp['field18'].apply(lambda x: 1 if x in [6,7] else 0)
        tmp['field20'] = tmp['field20'].apply(lambda x: 1 if x in [4,5,6] else 0)
        #encoding categorial features
        cat_features = ['field2', 'field3', 'field11', 
                        'field23', 'field15', 'field24', 'field29']
        ohe = OneHotEncoder(sparse=False)
        ohe.fit(self.X[cat_features])
        encoded_features = pd.DataFrame(ohe.transform(tmp[cat_features]))
        dummy_cols_names = []
        for i in range(len(cat_features)):
            tmp1 = [f'{cat_features[i]}_{cat}' for cat in ohe.categories_[i]]
            dummy_cols_names += tmp1
        
        encoded_features.columns = dummy_cols_names
        encoded_features.set_index(tmp.index, inplace = True)
        tmp.drop(cat_features, axis=1, inplace=True)
        tmp = pd.concat([tmp, encoded_features], axis=1)
        return tmp
```
